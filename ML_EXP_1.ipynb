```python
#SCENARIO 2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path=r"/content/diabetes.csv";
data=pd.read_csv(file_path,encoding="latin1")
print("POOVIKA M\n24BAD087\n")

print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
print(data.isnull())
print(data.columns)

# Visualization: Glucose

plt.figure(figsize=(10,6))
sns.histplot(data['Glucose'], bins=30, kde=True)
plt.title("Glucose Level Distribution (Histogram)")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=data['Glucose'])
plt.title("Glucose Level Distribution (Boxplot)")
plt.xlabel("Glucose Level")
plt.show()

# Visualization: Age

plt.figure(figsize=(10,6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title("Age Distribution (Histogram)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=data['Age'])
plt.title("Age Distribution (Boxplot)")
plt.xlabel("Age")
plt.show()

# Identify simple patterns

# Correlation between health metrics
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Health Metrics")
plt.show()
```

```python
#SCENARIO 3

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path=r"/content/Housing.csv"
my_data=pd.read_csv(file_path,encoding="latin1")
print("POOVIKA M\n24BAD087\n")

print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)
plt.figure(figsize=(7,5))
plt.scatter(my_data['area'], my_data['price'], color='blue')
plt.title("Price vs Area")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(my_data['bedrooms'], my_data['price'], color='green')
plt.title("Price vs Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(my_data['bathrooms'], my_data['price'], color='red')
plt.title("Price vs Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.show()


plt.figure(figsize=(10,8))
sns.heatmap(my_data[['price','area','bedrooms','bathrooms','stories','parking']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Housing Features")
plt.show()
```

```python
#SCENARIO 4

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path=r"/content/marketing_campaign.csv"
my_data=pd.read_csv(file_path,encoding="latin1",sep="\t")
print("POOVIKA M\n24BAD087\n")

print(my_data.head())
print(my_data.tail())
print(my_data.info())
print(my_data.describe())
print(my_data.isnull())
print(my_data.columns)
my_data['Age'] = 2026 - my_data['Year_Birth']

# BAR PLOTS

# Age bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['Age'].value_counts().index,
            y=my_data['Age'].value_counts().values)
plt.title("Age Distribution")
plt.show()

# Income bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['Income'].value_counts().index[:10],
            y=my_data['Income'].value_counts().values[:10])
plt.title("Income Distribution")
plt.show()

# Spending bar plot (Wine)
plt.figure(figsize=(6,4))
sns.barplot(x=my_data['MntWines'].value_counts().index[:10],
            y=my_data['MntWines'].value_counts().values[:10])
plt.title("Spending Pattern (Wine)")
plt.show()

# BOX PLOTS

# Age box plot
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['Age'])
plt.title("Age Box Plot")
plt.show()

# Income bar plot
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['Income'])
plt.title("Income Box Plot")
plt.show()

# Spending bar plot (Wine)
plt.figure(figsize=(5,3))
sns.boxplot(x=my_data['MntWines'])
plt.title("Spending Box Plot")
plt.show()
```
